### Startpunkte definieren

Das Timegrid wird über die Monatsebene angesprochen. Es werden jeweils die ersten vier Tage des Monats ausgewählt und die verknüpften Trade-Datensätze des Basiswerts `c` geladen. An Wochendenden wird nichts zurückgegeben. Es wird der erste gültige Datensatz ausgewählt.
Abschließend wird die Ergebnismenge nach Datum sortiert.

```ruby
v= Tg::Monat[12].tag( 1)
                .map{|y| y.vector(4)
                          .map{|y| y.ohlc_bar( c ){ TG::HasTrades } }
                          .compact.first }.compact
dez_start =  v.sort{|y,x| y.datum <=> x.datum }
```

Laden der November und Dezember Graphen

```ruby
nov_start =  dez_start.map{ |x| x.vector(-26).reverse} 
dez_end =  dez_start.map{ |x| x.vector(20)} 

```

Monatsdifferenzen November

```ruby
n0 = Polars::DataFrame.new( nov_start.map {|y| { datum: y.first.datum.year,
                           nov: (y.last.close - y.first.close) / y.first.close * 100 }} )
d1= Polars::DataFrame.new( dez_end.map{|y|  { datum: y.first.datum.year,
                             dez_w1: (y.at(3).close - y.first.close)/ y.first.close * 100 }})
d2= Polars::DataFrame.new( dez_end.map{|y|  { datum: y.first.datum.year,
                             dez_w2: (y.at(10).close - y.at(5).close)/ y.first.close * 100 }})
d3= Polars::DataFrame.new( dez_end.map{|y|  { datum: y.first.datum.year,
                             dez_w3: (y.at(15).close - y.at(10).close)/ y.first.close * 100 }})

n = n0.join( d1, on: 'datum', how: 'outer')
      .join( d2, on: 'datum', how: 'outer')
      .join( d3, on: 'datum', how: 'outer') 
```

```ruby
woche =  1
Vega.lite 
   .data(n)
.title( text: "November + Dezember Performance ",
         color: "navy", fontSize: 22 , align: "center", 
        anchor: "middle", dy: -1, baseline: "middle",
  	  subtitle: "#{c.symbol}  (normiert)",
 subtitleColor: "navy", subtitleFontSize: 18, subtitlePadding: 5)
   .layer( [
      { mark: { type: "bar" },
    encoding: { x: { field: "datum", type: "nominal"},
                y: { field: "nov", type: "quantitative",
                      axis: { title: "November Performance",
                             orient: 'left', 
                               grid: true,
                         titleColor: { value: '#b45309' } } },
      color: { value: '#b45309' } } }, # amber700
      { mark: {type: "bar" },
    encoding: { x: { field: "datum", type: "nominal",
             axis: { title: "Woche #{woche} "}   },
                y: { field: "dez_w#{woche}", type: "quantitative",
                     axis: { title: "Dezember Performance",
                            orient: 'right', 
                              grid: true,
                        titleColor: { value: '#0f766e'}}},
              color: { value: '#0f766e' } } }  # teal700
     ] )
```
