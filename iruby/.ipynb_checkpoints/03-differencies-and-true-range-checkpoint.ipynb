```ruby
spx = Contract.first
```

### Anfangs- und Enddaten sammeln

```ruby
start_dates = Ohlc.start_entry(spx).map &:datum
end_dates = Ohlc.last_entry(spx).map &:datum
trades, vola = start_dates.zip end_dates

IRuby.display IRuby.html ["<h3>Start und Enddatenpunkte </h3>",
                 "<tt>trades: #{ trades.join " ---> "}</tt> ",
                 "<tt>vola:   #{ vola.join " ---> "} </tt>"] 
```

### Freitage als Startpunkte definieren

```ruby
select_all_fridays = ->(v){ (v).select{|x| x.wday == 5}}

all_fridays= select_all_fridays[ Vola.start_entry(spx).first.datum .. Vola.last_entry(spx).first.datum  ]

IRuby.display IRuby.html "<b>all_fridays</b> enthält <b>#{all_fridays.size}</b> Datensätze" 
```

### Für alle Datensätze die Preisdifferenz von 28 Tagen ermitteln. 
(Bezug: close)

```ruby
differences = all_fridays.map do | f |
  start_node = f.to_tg.ohlc_bar( spx ){ Tg::HasTrades }
  vola = f.to_tg.ohlc_bar( spx ){ Tg::HasVola }
  next if start_node.nil? || vola.nil?
  last_node = (f+28).to_tg.ohlc_bar( spx ){ Tg::HasTrades }
  next if last_node.nil?
  {       date: f, 
    difference: last_node.close - start_node.close, 
          vola: vola.close, 
     norm_diff: (last_node.close - start_node.close)/ start_node.close ,
          open: start_node.close
  }
end.compact
IRuby.display IRuby.table differences
```

```ruby
Vega.lite 
  .data( differences )
  .title( text: "28DTE Kursdifferenzen ",
         color: "navy",
        fontSize: 22 ,
  		 align: "center",
  		anchor: "middle",
  			dy: -1,
      baseline: "middle",
  	  subtitle: "S&P 500  (normiert)",
 subtitleColor: "navy", 
subtitleFontSize: 18,
subtitlePadding: 5)
  .resolve( scale: { y: "independent"})
  .layer( [ 
         { mark:{ type: "line", interpolate: "cardinal", color: '#c2410c'},
           encoding: { x:{ field: "date", type: "temporal",
                           title: '' },
                       y:{ field: 'norm_diff', type: 'quantitative',
                           title: 'Norm. Preisdifferenz 28 Tage',
                           axis: { orient: 'left',titleColor:  '#c2410c'},
                           scale: { zero: false} }           }
          },
          { mark:{ type: "line", interpolate: "cardinal", color: '#0d9488'},
           encoding: { x:{ field: "date", type: "temporal"},
                       y:{ field: 'vola', type: 'quantitative',
                           title: 'Implied Volatiltiy',
                           axis: { orient: 'right', titleColor:  '#0d9488'},
                           scale: { zero: false} }           }
          }
          ])
```

```ruby
s= Polars::DataFrame.new( Vola.start_entry(spx).vector(0).first.map do |y|
   { date: y.datum, ref_vola: y.wap} 
  end )
```

```ruby
d= differences.find_all{|y| y[:norm_diff] > 0.088 || y[:norm_diff] < -0.068 }
IRuby.display "Anzahl: #{d.size}"
d << { date: Date.today, min: -0.07, max: 0.09, difference: 0, vola: 0, norm_diff: 0}
IRuby.display IRuby.table d
```

```ruby
t = Polars::DataFrame.new differences
```

Bei einem Stangle mit Delta(init) von 0,2 spannt sich ein Toleranzbereich von aktuell 380 Punkten (normiert 0,088) auf. Da die Call-Seite weniger Prämie erwirtschaftet, muss hier ein engeres Risikomaß verwendet werden

```ruby
st = s.join( t, on: 'date', how: 'outer')
```

```ruby
Vega.lite
  .title( text: "Ausreißer, statische Betrachtung ",
         color: "navy",
        fontSize: 22 ,
  		 align: "center",
  		anchor: "middle",
  			dy: -1,
      baseline: "middle",
  	  subtitle: "S&P 500  (normiert)",
 subtitleColor: "navy", 
subtitleFontSize: 18,
subtitlePadding: 5)
  .data( st )
  .params( [{ name: "call", value: 0.08, 
              bind: { input: "range", min: 0.05, max: 0.1, step: 0.001 } },
            { name: "put", value: 0.08, 
              bind: { input: "range", min: 0.05, max: 0.1, step: 0.001 } },
            { name: "barcolor",  value: "#10b981" },
            { name: "assignedcolor",  value: "#ef4444" },
            { name: "linecolor", value: "#d6d3d1" }
           ])
  .vconcat( [ {
     resolve: { scale: { y: "independent"} },
     width: 650,
     layer:[ { mark: { type: "bar" } ,
            
           encoding: { opacity: { value: 0.8 },
                             x: { field: "date", 
                                   type: "temporal",
                               timeunit: "day",
                                  scale: { domain: { param: "brush" } }   
                                 },
                             y: { field: "norm_diff", type: "quantitative",
                                   axis: { title: "Monatliche Preisdifferenz (norm.)",
                                          orient: 'left', 
                                            grid: true,
                                      titleColor:  { expr: "assignedcolor"}}},
                            color:  {  condition: { test: "datum['norm_diff'] < (put * -1) || 
                                                          datum['norm_diff'] > call" ,
                                                   value: { expr: "assignedcolor"}},
                                           value: { expr: "barcolor"} }     
                                 
                                  }
                     
              } ,
           { mark: { type: "line", interpolate: 'cardinal', color: { expr: "linecolor" }},
                 encoding: { y: { field: "ref_vola", type: 'quantitative',
                              axis: { orient: 'right', 
                                  titleColor: { expr: "linecolor"},
                                      title: "Implied Volatiltiy"},
                                      color: { value: { expr: "linecolor"} } },
                        x: {field: "date", type: "temporal", timeunit: "day",
                            scale: { domain: { param: "brush" } } }
                           }
               } ] 
             } ,
           { 
      width: 650,
     height: 60,
    opacity: { value:0.7 },
       mark: { type: "bar", width: 1},
     params: [ { name: "brush",
                select: { type: "interval", encodings: ["x"] } }],
   encoding: { x: {field: "date", type: "temporal", timeunit: "day"},
               y: {field: "norm_diff", type: "quantitative",
                    axis: { title: "Events"} },
           color:  {  condition: { test: "datum['norm_diff'] < (put * -1) || 
                                          datum['norm_diff'] > call" ,
                                          value: { expr: "assignedcolor"}},
                                   value: { expr: "barcolor"} }
              } 
            } ]
  )
  
```
